# Interactive autodidactic school
#### A new metaheuristic optimization algorithm for solving mathematical and structural design optimization problems

This is a simple python implementation of the Interactive Autodidactic School algorithm recently proposed by M. Jahangiri et al. in in the paper entitled *Interactive autodidactic school: A new metaheuristic optimization algorithm for solving mathematical and structural design optimization problems*. https://www.sciencedirect.com/science/article/abs/pii/S0045794920300717

Based on the paper, IAS is a population-based algorithm on the basis of the interactions between students in an autodidactic school with the goal of increasing their knowledge through a combination of self-teaching/self-learning, interactive discussion, criticism and the competition.

I've implemented the suggested procedure in a module named `InteractiveAutodidacticSchool`. So let's import it:

```python
import numpy as np
import InteractiveAutodidacticSchool as opt
import matplotlib.pyplot as plt
```

### Function definition
Below, some of the functions mentioned in the paper are defined and tested using the IAS implementation.

```python
def easom_func(x: np.ndarray):
    "Returns the Easom Function (Exact Min = -1) @ x=(pi, pi)"
    return (-np.cos(x[0])*np.cos(x[1])*np.exp(-(x[0]-np.pi)**2 - (x[1]-np.pi)**2))

x_b_easom = np.array([[-100,100],
                     [-100,100]
                    ])


def six_hump_camel_func(x: np.ndarray):
    """Returns the Six-Hump Camel function
       Exact Min = -1.0316 @ x=(-0.0898, -0.7126)
    """
    return (4*x[0]**2 - 2.1*x[0]**4 + (1/3)*(x[0]**6) +
            x[0]*x[1] - 4*x[1]**2 + 4*x[1]**4)

x_b_shc = np.array([[-5, 5],
                     [-5, 5]
                    ])


def booth_func(x: np.ndarray):
    "Returns the Booth function (Exact Min = 0) @ x=(1, 3)"
    return (x[0] + 2*x[1] - 7)**2 + (2*x[0] + x[1] - 5)**2

x_b_booth = np.array([[-10, 10],
                     [-10, 10]
                    ])


def sphere_func(x: np.ndarray):
    "Returns the Sphere function (Exact Min = 0 @ x={0})"
    return np.sum(x**2)

x_b_sphere = np.column_stack((-100*np.ones(30),100*np.ones(30)))


def schwefel2_func(x: np.ndarray):
    "Returns the Schwefel2 function (Exact Min = 0 @ x={0})"
    return np.sum(np.abs(x)) + np.prod(np.abs(x))

x_b_schwefel2 = np.column_stack((-10*np.ones(30),10*np.ones(30)))


def rastrigin_func(x: np.ndarray):
    "Return the Rastrigin function (Exact Min = 0 @ x={0})"
    return np.sum(x**2 - 10*np.cos(2*np.pi*x) + 10)

x_b_rastrigin = np.column_stack((-5.12*np.ones(30), 5.12*np.ones(30)))
```

### Optimization

```python
IAS_0 = opt.IAS(easom_func, x_b_easom, 10, 50)
IAS_0.do_optimization();
```

```python
IAS_1 = opt.IAS(six_hump_camel_func, x_b_shc, 10, 50)
IAS_1.do_optimization();
```

```python
IAS_2 = opt.IAS(booth_func, x_b_booth, 10, 50)
IAS_2.do_optimization();
```

```python
IAS_3 = opt.IAS(sphere_func, x_b_sphere, 30, 100)
IAS_3.do_optimization();
```

```python
IAS_4 = opt.IAS(schwefel2_func, x_b_schwefel2, 30, 100)
IAS_4.do_optimization();
```

```python
IAS_5 = opt.IAS(rastrigin_func, x_b_rastrigin, 30, 100)
IAS_5.do_optimization();
```

### Visualization

```python
fig, axs = plt.subplots(3, 2, figsize=(13,10))
for ax in axs.flat:
    ax.set_xlabel('Number of iterations',{'fontname':'Cambria', 'fontstyle':'italic','size':13})
    ax.set_ylabel('f(x)',{'fontname':'Cambria', 'fontstyle':'italic','size':13})
    ax.grid()
    
axs[0,0].plot(IAS_0.history['iters'], IAS_0.history['f(x)'])
axs[0,0].set_title('Convergence History of Easom Function')

axs[0,1].plot(IAS_1.history['iters'], IAS_1.history['f(x)'])
axs[0,1].set_title('Convergence History of Six-Hump Camel Function')

axs[1,0].plot(IAS_2.history['iters'], IAS_2.history['f(x)'])
axs[1,0].set_title('Convergence History of Booth function Function')

axs[1,1].plot(IAS_3.history['iters'], IAS_3.history['f(x)'])
axs[1,1].set_title('Convergence History of Sphere Function')

axs[2,0].plot(IAS_4.history['iters'], IAS_4.history['f(x)'])
axs[2,0].set_title('Convergence History of Schwefel2 Function')

axs[2,1].plot(IAS_5.history['iters'], IAS_5.history['f(x)'])
axs[2,1].set_title('Convergence History of Rastrigin Function')

fig.tight_layout()
```
